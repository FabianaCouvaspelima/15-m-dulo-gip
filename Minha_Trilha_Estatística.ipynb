GRÁFICO DE BARRAS
Cada categoria (eixo horizontal)é representada por uma barra cuja altura corresponde à frequência ou frequência relativa da categoria correspondente(eixo vertical). Entre barras adjacentes são deixadas lacunas tal como na figura a seguir.

```python
import matplotlib.pyplot as plt

x=[1, 2, 3, 4, 5, 6]
y=[50, 100, 200, 200, 100, 50]
plt.bar(x,y,color='blue')
plt.ylabel('Frequência')
plt.grid()
```

Utilizar os dados para construir uma tabela com as informações acerca da distribuição de frequências relativas.

```python
total = sum(y)
p = []
for yi in y:
  p.append(yi/total)
  print(p)
  p
```

```python
plt.bar(x,p,color='blue')
plt.grid()
```
